## **Efisiensi Interpolation Search Iteratif dan Rekursif**
Efisiensi algoritma pencarian interpolasi dapat diukur berdasarkan waktu eksekusi algoritma ketika mencari elemen pada berbagai skenario distribusi data. Dalam hal ini, kita menggunakan tiga jenis distribusi data:
1. **Uniform Distribution**: Data tersebar secara merata di rentang tertentu.
2. **Normal Distribution**: Data tersebar mengikuti kurva Gaussian (bell-shaped curve).
3. **Exponential Distribution**: Data tersebar mengikuti distribusi eksponensial.

Efisiensi algoritma diuji menggunakan pendekatan iteratif dan rekursif dengan ukuran input yang bervariasi:

### **Ukuran Input**
Ukuran input yang digunakan untuk pengujian meliputi:
- **1,000**
- **2,000**
- **5,000**
- **10,000**
- **20,000**
- **50,000**
- **100,000**
- **200,000**
- **500,000**
- **1,000,000**

### **Pengujian dan Basic Operation**
Untuk mengukur efisiensi algoritma pencarian interpolasi, kita memilih *basic operation* yang relevan, yaitu:
1. **Membandingkan elemen array** pada saat interpolasi (menghitung posisi yang diperkirakan elemen target).
2. **Input Size (n)**: Menyatakan jumlah elemen dalam array yang digunakan untuk pencarian.

Pengujian dilakukan dengan mengukur waktu eksekusi algoritma menggunakan dua pendekatan:
- **Iteratif**: Pendekatan berbasis loop.
- **Rekursif**: Pendekatan berbasis rekursi.

Hasil dari pengujian ini akan membantu kita memahami perbedaan performa antara pendekatan iteratif dan rekursif pada berbagai distribusi data dan ukuran input.


## **Perhitungan Kompleksitas Waktu Iteratif dan Rekursif**
### **Worst Case Scenario Iteratif dan Rekursif**

Pada *Interpolation Search*, skenario kasus terburuk terjadi ketika elemen yang dicari tidak berada di dalam array atau distribusi data tidak memungkinkan interpolasi yang akurat. Hal ini menyebabkan banyak operasi interpolasi gagal mendekati elemen target. Untuk algoritma iteratif dan rekursif, ini dapat terjadi pada distribusi data yang sangat tidak teratur atau memiliki nilai yang jauh berbeda dari pola distribusi ideal.

### **Analisis Kompleksitas Waktu**
Pada skenario kasus terburuk, waktu eksekusi *Interpolation Search* iteratif maupun rekursif mendekati **O(n)**. Hal ini terjadi karena:
- Setiap langkah interpolasi salah memperkirakan posisi elemen target, sehingga algoritma harus memindai data lebih banyak.
- Dalam skenario tertentu, *Interpolation Search* iteratif dan rekursif dapat mendekati *Linear Search* pada array yang terurut, di mana setiap elemen harus diperiksa secara langsung.

#### **Formulasi Kasus Terburuk**
Dalam kasus terburuk, kompleksitas waktu *Interpolation Search* dapat dirumuskan sebagai berikut:

$$
T(0) = 0
$$

$$
T(n) = T(n-1) + C(n)
$$

Dengan \(C(n)\) adalah waktu untuk melakukan interpolasi dan perbandingan elemen. Untuk distribusi data yang tidak ideal, \(C(n) \in O(n)\), sehingga:

$$
T(n) \in O(n)
$$

Kasus terburuk ini lebih mungkin terjadi pada array yang sangat besar dengan distribusi data yang tidak ideal.

#### **Rumus Kompleksitas**
Kompleksitas waktu untuk pendekatan iteratif dan rekursif pada kasus terburuk dirumuskan sebagai:

$$
T(n) = \frac{n(n-1)}{2} \in O(n)
$$

Jika elemen target berada di luar rentang array:
1. Pendekatan iteratif melakukan iterasi hingga batas array tercapai.
2. Pendekatan rekursif memanggil fungsi hingga mencapai kondisi dasar, menyebabkan tumpukan rekursi.

---

Kasus terburuk ini menunjukkan bahwa meskipun *Interpolation Search* sangat efisien pada data yang terdistribusi ideal, performa algoritma dapat menurun drastis dalam distribusi yang tidak teratur atau data dengan elemen pencarian yang tidak relevan.


### **Best Case Scenario Iteratif dan Rekursif**
Pada skenario terbaik, setiap langkah interpolasi mendekati posisi target dengan akurasi tinggi, sehingga pencarian selesai dengan jumlah langkah minimal. Untuk algoritma iteratif dan rekursif, ini terjadi jika data terdistribusi secara merata dan elemen target berada tepat di posisi yang diperkirakan oleh interpolasi.

### **Analisis Kompleksitas Waktu**
Pada skenario terbaik, waktu eksekusi *Interpolation Search* mendekati **O(\log(\log(n)))**. Hal ini terjadi karena:
- Setiap langkah interpolasi memperkirakan posisi target dengan sangat akurat, sehingga jumlah langkah pencarian sangat sedikit.
- Algoritma hanya perlu melakukan beberapa perbandingan sebelum menemukan elemen target.

#### **Formulasi Kasus Terbaik**
Dalam kasus terbaik, kompleksitas waktu dapat dirumuskan sebagai berikut:

$$
T(0) = 0
$$

$$
T(n) = T(n-1) + C(\log(n))
$$

Dengan \(C(\log(n))\) adalah waktu untuk melakukan interpolasi. Karena interpolasi sangat akurat pada skenario terbaik:

$$
T(n) \in O(\log(\log(n)))
$$

#### **Simulasi Kasus Terbaik**
- Pada **Uniform Distribution**, interpolasi langsung memperkirakan posisi elemen target secara akurat pada langkah pertama.
- Pada **Normal Distribution**, interpolasi berhasil menemukan elemen target di sekitar mean (\(\mu\)) dengan jumlah iterasi minimal.
- Pada **Exponential Distribution**, interpolasi cepat menemukan elemen target di awal distribusi dengan jumlah langkah yang kecil.

#### **Rumus Kompleksitas**
Kompleksitas waktu untuk pendekatan iteratif dan rekursif pada kasus terbaik dirumuskan sebagai:

$$
T(n) \approx 2 \ln(\ln(n))
$$

Hal ini menunjukkan bahwa waktu pencarian sangat efisien pada data yang terdistribusi ideal.

---

Skenario terbaik ini menggambarkan potensi optimal dari algoritma *Interpolation Search* untuk menyelesaikan pencarian dengan jumlah langkah minimal pada data yang terdistribusi secara merata.


## **Notasi Asimtotik Iteratif dan Rekursif**

### **Notasi Asimtotik Worst Case Scenario Iteratif dan Rekursif**
Pada skenario kasus terburuk, algoritma *Interpolation Search* akan mendekati performa **Linear Search** jika distribusi data sangat tidak ideal atau elemen target berada di luar rentang array. Hal ini membuat algoritma harus memproses elemen satu per satu. Kompleksitas waktu pada kasus terburuk dapat dirumuskan sebagai:

$$
T(n) \in O(n)
$$

### **Notasi Asimtotik Best Case Scenario Iteratif dan Rekursif**
Pada skenario terbaik, kompleksitas waktu dapat dirumuskan sebagai:

$$
T(n) \in O(\log(\log(n)))
$$

---

## **Class Efficiency Iteratif dan Rekursif**
Efisiensi kelas algoritma *Interpolation Search* dapat diklasifikasikan sebagai berikut:
1. **Class Efficiency Worst Case**: **Linear**

   $$ 
   O(n)
   $$

2. **Class Efficiency Best Case**: **Double Logarithmic**

   $$
   O(\log(\log(n)))
   $$

Efisiensi ini mencerminkan performa algoritma berdasarkan distribusi data dan posisi elemen target. Pada kasus terbaik, algoritma sangat cepat, mendekati posisi target hanya dalam beberapa langkah. Sebaliknya, pada kasus terburuk, algoritma memerlukan banyak iterasi untuk mencapai hasil.


```python
import time
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
import math
from matplotlib.ticker import ScalarFormatter, LogLocator
```

```python
def interpolation_search_iterative(arr, x):
    low = 0
    high = len(arr) - 1

    while low <= high and x >= arr[low] and x <= arr[high]:
        if low == high:
            if arr[low] == x:
                return low
            return -1

        pos = low + int(((float(high - low) / (arr[high] - arr[low])) * (x - arr[low])))

        if arr[pos] == x:
            return pos
        if arr[pos] < x:
            low = pos + 1
        else:
            high = pos - 1

    return -1
```

```python
def interpolation_search_recursive(arr, low, high, x):
    if low > high or x < arr[low] or x > arr[high]:
        return -1

    if low == high:
        if arr[low] == x:
            return low
        else:
            return -1

    pos = low + int(((float(high - low) / (arr[high] - arr[low])) * (x - arr[low])))

    if arr[pos] == x:
        return pos
    elif arr[pos] < x:
        return interpolation_search_recursive(arr, pos + 1, high, x)
    else:
        return interpolation_search_recursive(arr, low, pos - 1, x)
```

```python
np.random.seed(42)

def generate_data_uniform(n, low=1, high=10**6):
    data = np.random.randint(low, high + 1, size=n)
    return np.sort(data)

def generate_exponential_data(n, scale=1.0):
    data = np.random.exponential(scale=scale, size=n)
    return np.sort(data)

def generate_normal_data(n, mean=0, std_dev=1.0):
    data = np.random.normal(loc=mean, scale=std_dev, size=n)
    return np.sort(data)
```

```python
def measure_time(search_func, arr, x, iterations=5):
    total_time = 0
    idx = -1

    for _ in range(iterations):
        start = time.perf_counter()
        if search_func.__name__ == "interpolation_search_recursive":
            idx = search_func(arr, 0, len(arr)-1, x)
        else:
            idx = search_func(arr, x)

        end = time.perf_counter()
        total_time += (end - start) * 1000

    avg_elapsed = total_time / iterations
    return avg_elapsed, idx
```

```python
sizes = [1_000, 2_000, 5_000, 10_000, 20_000, 50_000, 100_000, 200_000, 500_000, 1_000_000]

results = {
    'uniform': {
        'iterative_time': [],
        'recursive_time': []
    },
    'normal': {
        'iterative_time': [],
        'recursive_time': []
    },
    'exponential': {
        'iterative_time': [],
        'recursive_time': []
    }
}
```

```python
for n in sizes:
    np.random.seed(42)
    print(f"Mengukur waktu pada n = {n} ...")

    data_u = generate_data_uniform(n)
    x_u = data_u[n // 2]

    time_iter_u, _ = measure_time(interpolation_search_iterative, data_u, x_u)
    results['uniform']['iterative_time'].append(time_iter_u)

    time_rec_u, _ = measure_time(interpolation_search_recursive, data_u, x_u)
    results['uniform']['recursive_time'].append(time_rec_u)

    data_n = generate_normal_data(n, mean=0, std_dev=1.0)
    x_n = data_n[n // 2]

    time_iter_n, _ = measure_time(interpolation_search_iterative, data_n, x_n)
    results['normal']['iterative_time'].append(time_iter_n)

    time_rec_n, _ = measure_time(interpolation_search_recursive, data_n, x_n)
    results['normal']['recursive_time'].append(time_rec_n)

    data_e = generate_exponential_data(n, scale=1.0)
    x_e = data_e[n // 2]

    time_iter_e, _ = measure_time(interpolation_search_iterative, data_e, x_e)
    results['exponential']['iterative_time'].append(time_iter_e)

    time_rec_e, _ = measure_time(interpolation_search_recursive, data_e, x_e)
    results['exponential']['recursive_time'].append(time_rec_e)

    print(f"Selesai mengukur n = {n}\n")
```

```python
def plot_results(sizes, results):
    data = []
    for distribution, times in results.items():
        for i, size in enumerate(sizes):
            data.append({
                "Ukuran Data": size,
                "Waktu Eksekusi (ms)": times["iterative_time"][i] * 1000,
                "Kategori": f"{distribution.capitalize()} Iterative"
            })
            data.append({
                "Ukuran Data": size,
                "Waktu Eksekusi (ms)": times["recursive_time"][i] * 1000,
                "Kategori": f"{distribution.capitalize()} Recursive"
            })

    df = pd.DataFrame(data)

    plt.figure(figsize=(12, 7))
    sns.lineplot(
        data=df,
        x="Ukuran Data",
        y="Waktu Eksekusi (ms)",
        hue="Kategori",
        marker="o"
    )

    for _, row in df.iterrows():
        plt.plot([row["Ukuran Data"], row["Ukuran Data"]],
                 [0, row["Waktu Eksekusi (ms)"]],
                 color='gray', linestyle='--', linewidth=0.5)

    plt.xlabel('Ukuran Data (n)', fontsize=14)
    plt.ylabel('Waktu Eksekusi (ms)', fontsize=14)
    plt.title('Perbandingan Waktu Eksekusi Interpolation Search', fontsize=16)

    plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', fontsize=12, title="Kategori")

    plt.grid(True, which="both", linestyle="--", linewidth=0.5)

    plt.xscale('log')

    plt.tight_layout()
    plt.show()
```

```python
plot_results(sizes, results)
```

- **Iteratif**: Distribusi eksponensial menunjukkan waktu eksekusi yang meningkat signifikan pada ukuran input besar. Contohnya: **325 ms untuk 500.000 elemen**.

**Kesimpulan**: 
- Pada distribusi eksponensial, elemen target di ekor distribusi memerlukan lebih banyak langkah interpolasi. Hal ini menyebabkan waktu eksekusi lebih lama dibandingkan distribusi uniform dan normal.

---

### **5. Perbandingan Iteratif dan Rekursif**
- **Pendekatan Iteratif**: Cenderung lebih cepat pada semua jenis distribusi, terutama untuk ukuran data besar.
- **Pendekatan Rekursif**: Memiliki performa yang kompetitif dengan iteratif, tetapi waktu eksekusinya sedikit lebih tinggi pada distribusi yang sulit seperti eksponensial.

---

### **6. Tren dan Efisiensi**
- **Uniform**: Performa terbaik, dengan waktu eksekusi paling rendah di semua ukuran input.
- **Normal**: Waktu eksekusi lebih tinggi daripada uniform tetapi tetap lebih baik dibandingkan eksponensial.
- **Exponential**: Distribusi ini memiliki performa terburuk, terutama pada ukuran input yang besar, karena elemen di ekor distribusi membutuhkan lebih banyak langkah interpolasi.
